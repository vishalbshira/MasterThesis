droping uncesssary and redundant columns

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1', 'host_since', 'host_response_time', 'neighbourhood_cleansed', 'country_code', \
         'latitude', 'longitude', 'availability_30', 'availability_365'], axis=1, inplace=True)
```

lets convert required columns as categorical datatype

```python
for col in ['host_response_rate', 'market', 'property_type', 'room_type', 'cancellation_policy']:
    df[col] = df[col].astype('category')
```

lets convert required columns as boolean datatype

```python
for col in ['host_is_superhost', 'host_identity_verified', 'instant_bookable', 'air_conditioning', 'balcony', 'bed_linen', 'tv', 'coffee_machine',
       'cooking_basics', 'white_goods', 'elevator', 'child_friendly',
       'parking', 'host_greeting', 'hot_tub_sauna_or_pool', 'internet', 'long_term_stays', 'pets_allowed', 'private_entrance', 
        'smoking_allowed']:
    df[col] = df[col].astype('bool')
```

```python
df.info()
```

seting target and features

```python
target = df[['price']]

features = df.drop(['price'], axis = 1)
```

encoding the categorical variables

```python
features.select_dtypes(include=['float64', 'int64', 'bool']).head()
```

```python
numerical_features = features.select_dtypes(include=['float64', 'int64', 'bool']).copy()

categorical_features = features.select_dtypes(include=['category']).copy()
categorical_features = pd.get_dummies(categorical_features)
```

```python
encoded_features = pd.concat([numerical_features, categorical_features], axis=1)
```

Split the data into traning and testing datesets

```python
# used for splitinf the dataset
from sklearn.model_selection import train_test_split

# used for evaluating the models
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(encoded_features, target, test_size = 0.25, random_state= 0)
```
